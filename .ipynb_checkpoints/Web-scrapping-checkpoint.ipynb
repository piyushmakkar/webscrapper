```python
#importing libraries

from bs4 import BeautifulSoup
import requests
```

```python
#loading the page

r = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/#t=0&s=0", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c = r.content

soup = BeautifulSoup(c,"html.parser")
```

```python
#iterating to get the values
#and storing in list via dictionary

l=[]
all=soup.find_all("div",{"class":"propertyRow"})
for items in all:
    d={}
    
    d["Address"] = items.find_all("span",{"class":"propAddressCollapse"})[0].text
    d["Locality"] = items.find_all("span",{"class":"propAddressCollapse"})[1].text
    d["Price"] = items.find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")
    
    try:
        d["Beds"] = items.find("span",{"class":"infoBed"}).find("b").text
    except:
        d["Beds"] = None
    
    try:
        d["Sq feet"] = items.find("span",{"class":"infoSqFt"}).find("b").text
    except:
        d["Sq feet"] = None
    
    try:
        d["Full Baths"] = items.find("span",{"class":"infoValueFullBath"}).find("b").text
    except:
        d["Full Baths"] = None
    
    try:
        d["Half Baths"] = items.find("span",{"class":"infoValueHalfBath"}).find("b").text
    except:
        d["Half Baths"] = None
        
    for col_grp in items.find_all("div",{"class":"columnGroup"}):
        for feature_grp , feature_name in zip(col_grp.find_all("span",{"class":"featureGroup"}),col_grp.find_all("span",{"class":"featureName"})):
            if "Lot Size" in feature_grp.text:
                d["Lot Size"] = feature_name.text
    l.append(d)
```

```python
#Saving extracted data and creating data frames
import pandas as pd
df = pd.DataFrame(l)
```
